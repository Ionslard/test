# Hand Tracking Music Controller Demo

This notebook demonstrates how to use the `HandMusicController` to track your hand with a webcam and generate MIDI notes. The video feed will be displayed directly in the notebook.

## Step 1: Setup a Virtual MIDI Port

Before running the code, you need a way to hear the MIDI notes. This requires two pieces of software:

1.  **A virtual MIDI driver:** This creates a virtual MIDI cable on your computer. The `HandMusicController` will send notes to this virtual cable.
    *   **Windows:** [loopMIDI](https://www.tobias-erichsen.de/software/loopmidi.html)
    *   **macOS:** The *Audio MIDI Setup* utility is built-in. You can create a virtual port in the IAC Driver.

2.  **A synthesizer/DAW:** This software will listen for notes on the virtual MIDI port and play sounds.
    *   **Examples:** GarageBand (macOS), Ableton Live, FL Studio, or a simple standalone synth like [SynthFont](https://www.synthfont.com/) (Windows).

**Instructions:**

1.  Install and run your virtual MIDI driver. Create a new port. The default name `HandTrackingMidiController` will work automatically.
2.  Open your synthesizer/DAW and set its MIDI input to the virtual port you just created.

## Step 2: Import and Initialize the Controller

First, we import the `HandMusicController` class from our module.

```python
from hand_tracking_module import HandMusicController
import time
```

Now, create an instance of the controller. If you named your virtual MIDI port something different, you can pass it as an argument here (e.g., `HandMusicController(midi_port_name='MyMIDIPort')`).

```python
controller = HandMusicController()
```

## Step 3: Start Hand Tracking

The `start_notebook_display()` method will activate the webcam and start showing the video feed below. The hand tracking and MIDI note generation will begin immediately.

When you are finished, run the `stop()` method. This will turn off the webcam, release all resources, and send a final 'note off' message.
